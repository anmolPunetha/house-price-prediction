```python
import numpy as np
import pandas as pd

import matplotlib as mp
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation 

from sklearn.datasets import load_boston
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from IPython.display import HTML 
#displaying the result
```

```python
#load the datasheet
boston=load_boston()

#description of dataset
print(boston.DESCR)
```

```python
#store data in pandas dataframes
features=pd.DataFrame(boston.data,columns=boston.feature_names)
features
```

```python
target=pd.DataFrame(boston.target,columns=['target'])
target
```

```python
max(target['target'])
```

```python
#concatenate features and target in a single dataframe
#bcz axis=1 se columnwise concatenate hoga
df=pd.concat([features,target], axis=1)
df
```

```python
df.describe().round(decimals=2)
```

```python
#CALCULATE CORRELATION between every column by using corr function of pd
corr=df.corr('pearson')

#take absolute values of correlation
corrs=[abs(corr[attr]['target']) for attr in list(features) ]

#make a list of pairs[corr,feature]
l=list(zip(corrs,list(features)))

#sort the list of pairs in reverse/descending order
#with the correlation value as the key for sorting
l.sort(key=lambda x:x[0], reverse=True)

#unzip pair to two lists
#zip(*l)- takes a list that looks like [[a,b,c],[d,e,f],[g,h,i]]
#and returns [[a,d,g],[b,e,h],[c,f,i]]
corrs,labels=list(zip((*l)))

#plot correlation wrt target 
index=np.arange(len(labels))
plt.figure(figsize=(15,5))
plt.bar(index,corrs,width=0.5)
plt.xlabel('Attributes')
plt.ylabel('Correlation with the targets')
plt.xticks(index,labels) #x-axis pe ticks
plt.show()
```

```python
#prep data for linear regression.....performing normalisation(o-1)
X=df['LSTAT'].values
Y=df['target'].values
x_scaler=MinMaxScaler()
X=x_scaler.fit_transform(X.reshape(-1,1))
X=X[:,-1]
y_scaler=MinMaxScaler()
Y=y_scaler.fit_transform(Y.reshape(-1,1))
Y=Y[:,-1]
print(Y[:5])
```

```python
def error(m,x,c,t): #x is attribute and t is target
    N=x.size
    e=sum(((m*x+c)-t)**2)
    return e*1/(2*N)
    
```

```python
#0.2 indicates 20 percent of the data is randomly sampled as testing data
# 2-2 features, target k train/test
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2)
```

```python
#yahan se we start with linear regression buildup
def update(m,x,c,t,learning_rate):
    grad_m=sum(2*((m*x+c)-t)*x)
    grad_c=sum(2*((m*x+c)-t))
    m=m-grad_m*learning_rate
    c=c-grad_c*learning_rate
    return m,c
```

```python
def gradient_descent(init_m,init_c,x,t,learning_rate,iterations,error_threshold):
    #init manje initial
    m=init_m
    c=init_c
    error_values=list()
    mc_values=list()
    for i in range(iterations):
        e=error(m,x,c,t)
        if e<error_threshold:
            print("Error less than threshold. Stopiing descent gradient")
            break
        error_values.append(e)
        m,c=update(m,x,c,t,learning_rate)
        mc_values.append((m,c))
    return m,c,error_values,mc_values
    
```

```python
%%time 
#called as magic function...from ipython
init_m=0.9
init_c=0
learning_rate=0.001 # keep below 0.025 to avoid overfit
iterations=250
error_threshold=0.001
m,c,error_values,mc_values=gradient_descent(init_m,init_c,xtrain,ytrain,learning_rate,iterations,error_threshold)
```

```python
#as no. of iterations are more, changes are less noticable
#to reduce proceswing time, it is advised to choose small values
mc_values_anim=mc_values[0:250:5]
```

```python
fig,ax=plt.subplots()
ln, =plt.plot([],[],'ro-',animated=True)

def init():
    plt.scatter(xtest,ytest,color='g')
    ax.set_xlim(0,1.0)
    ax.set_ylim(0,1.0)
    return (ln,)

def update_frame(frame):
    m,c=mc_values_anim[frame]
    x1,y1=-0.5,m* -0.5 +c
    x2,y2=1.5,m* 1.5 +c
    ln.set_data([x1,x2],[y1,y2])
    return (ln,)

anim=FuncAnimation(fig, update_frame,frames=range(len(mc_values_anim)),init_func=init,blit=True)

HTML(anim.to_html5_video())
```

```python
plt.scatter(xtrain,ytrain,color='b')
plt.plot(xtrain,(m*xtrain+c),color='r')
```

```python
plt.plot(np.arange(len(error_values)),error_values)
plt.xlabel("Iterations")
plt.ylabel("Error")
```

```python
#calculate the predictions on the test as a vectorised operation
predicted=(m*xtest)+c
```

```python
#calculating the MSE for the predicted values on the testing set
mean_squared_error(ytest,predicted)
```
